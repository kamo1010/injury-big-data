```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
from sklearn.metrics import classification_report
from sklearn.utils.class_weight import compute_class_weight
from sklearn.model_selection import cross_val_score, StratifiedKFold, train_test_split
from sklearn.feature_selection import SelectFromModel
from sklearn.preprocessing import StandardScaler, OneHotEncoder, PowerTransformer, PolynomialFeatures
from category_encoders import WOEEncoder, BinaryEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score, ConfusionMatrixDisplay, classification_report
```

```python
df = pd.read_csv("sample_data/labels.csv", sep=",", encoding= 'utf-8')
df = df.loc[:, df.isna().sum()<712]
df.drop(columns=['연번', '비고', '일'], inplace=True)

df.rename(columns = {'년도':'year', '월':'month', '일':'day', '평일/주말': 'day_of_week', '주야간':'day_or_night', '시간대':'period'}, inplace = True)
df.rename(columns = {'호기':'vehicle_number', '기상\r\n조건':'weather', '임무\r\n지시':'order', '상급부대': 'location', '환자신분':'patient_category', '환자\r\n증상':'injury'}, inplace = True)
df.rename(columns = {'환자유형 (외상 / 골절 / 심장 / 뇌 / 화상 / 기타)':'injury_type', '접수/이륙':'registration_takeoff', '출동시간':'departure_delay', '환자 인수지역 (PZ)': 'pz', 'PZ 도착/이륙':'pz_arrival_takeoff', '환자 후송지역 (LZ)':'lz'}, inplace = True)
df.rename(columns = {'LZ도착':'lz_arrival', '복귀이륙/복귀도착':'takeoff_arrival_at_basement', '출동~병원':'transfer_delay', '후송결과': 'y', '기상악화':'weather_deterioration', '호이스트': 'hoist'}, inplace = True)

df.head()
```

```python
def fit_y(x):
  x = str(x).replace("\r\n", "")
  x = x.replace(" ", "")
  if '생존' in x: x = 0
  elif '사망' in x: x = 1
  return pd.Series([x])

df.dropna(subset=['y', 'pz_arrival_takeoff', 'takeoff_arrival_at_basement'], inplace=True)
df[['y']] = df.y.apply(lambda x:fit_y(x))[df.y.isin(['생존', '사망'])]
df['y'].value_counts()

df.period = df.period.apply(lambda x:x.replace(" ~ ", "~"))
df.replace({"00~07": "00~06", "12~19": "12~18"}, inplace=True)
df['period'].value_counts()

df.fillna(value={'vehicle_number':0.}, inplace=True)

df.replace({"00~07": "00~06", "12~19": "12~18"}, inplace=True)
df['period'].value_counts()

df.weather = df.weather.apply(lambda x:x.replace(' ', ''))
df['weather'].value_counts()
df['weather'] = df['weather'].map({'NO-GO':0, 'GO':1, 1:1, 0:0})

df['day_or_night'] = df['day_or_night'].map({'주간':1, '야간':0, 1:1, 0:0})

def fit_order(x):
  x.replace(' ', '')
  x.replace('예비', '')
  if ('/' in x):
    return pd.Series(x.split('/'))
  return pd.Series([x, '없음'])
df[['order', 'change']] = df.order.apply(lambda x:fit_order(x))

def fit_patient_category(x):
  if x in ['-', '???']:
    x.replace(x, '?')
  if x in ['장교', '부사관', '준사관']:
    x.replace(x, '간부')
  if x in ['장교\r\n병' '간부, 병', '간부/병']:
    x.replace(x, '간부/병')
    return pd.Series(x.split('/'))
  return pd.Series([x])
df[['patient_category', 'company']] = df.patient_category.apply(lambda x:fit_patient_category(x))
tmp = df.loc[df['company']=='병']
tmp['patient_category'] = tmp['company']
pd.concat([df, tmp])
df.drop(columns=['company'], inplace=True)

def fit_injury_type(x):
  categories = ["외상", "골절", "심장", "뇌", "호상", "절단", "의식저하", "폐", "기타"]
  for val in categories:
    if val in x:
      x = val
  if x in [""]:
    x = "뇌"
  if x not in categories:
    x = x
  return pd.Series([x])

df[['injury_type']] = df.injury_type.apply(lambda x:fit_injury_type(x))
df['injury_type'].unique()
#df.loc[df['injury_type'].contains("정단")]

def fit_registration_takeoff(x):
  x = x.replace("\r\n", "")
  x = x.replace(" ", "")
  x = x.replace("(용인)", "")
  if len(x) == 4: x = x+"/"+x
  if len(x) == 8: x = x[:4]+"/"+x[4:]
  if len(x) == 13: x = x[:4]+"/"+x[9:]
  if x == "1757/1618": x = "1557/1618"
  if x == "2250/2058": x = "2250/2358"
  if x == "1159/0137": x = "2359/0137"
  return pd.Series([x])
df[['registration_takeoff']] = df.registration_takeoff.apply(lambda x:fit_registration_takeoff(x))
df['registration_takeoff'].unique()

def diff_delay(x):
  tmp = x.split("/")
  if tmp[1] == '000-': return pd.Series([x])
  h = int(tmp[1][:2])-int(tmp[0][:2])
  if h < 0: h = 24+h
  r = h*60+(int(tmp[1][2:])-int(tmp[0][2:]))
  if r < 0: r = 1440+r
  return pd.Series([r])

df[['departure_delay']] = df.registration_takeoff.apply(lambda x:diff_delay(x))
df['departure_delay'].unique()

def fit_pz(x):
  x = x.split("(")[0]
  x = x.replace("\r\n ", "\r\n")
  x = x.replace("\r\n", " ")
  x = x.replace(",", "")
  x = x.replace(" ", "")
  x = x.replace("패드", " PAD")
  x = x.replace(" PAD", "PAD")
  return pd.Series([x])
df[['pz']] = df.pz.apply(lambda x:fit_pz(x))
df['pz'].value_counts()

def fit_pz_arrival_takeoff(x):
  x = x.replace("\r\n", "")
  x = x.replace(" ", "")
  if len(x) == 4: x = x + "/" + x
  if len(x) == 5:
    if x[0] == "/": x = x[1:] + x
    else: x = x + x[:4]
  return pd.Series([x])
df[['pz_arrival_takeoff']] = df[df.pz_arrival_takeoff.notnull()].query("pz_arrival_takeoff != '-'").query("pz_arrival_takeoff != '×'").pz_arrival_takeoff.apply(lambda x:fit_pz_arrival_takeoff(x))
df['pz_arrival_takeoff'].unique()

def fit_lz(x):
  x = str(x).replace("\r\n", " ")
  return pd.Series([x])

df = df[df.lz.notnull()].query("lz != '-'")
df[['lz']] = df.lz.apply(lambda x:fit_lz(x))
df['lz'].value_counts()

def fit_lz_arrival(x):
  x = str(x).replace("\r\n", "")
  x = x.replace("/", "")
  x = x.replace("`", "")
  if x == '2:54': x = '0254'
  for i in range(4-len(x)):
    x = '0'+x
  return pd.Series([x])
df[['lz_arrival']] = df.query("lz_arrival != '×'").lz_arrival.apply(lambda x:fit_lz_arrival(x))
df['lz_arrival'].unique()

def fit_takeoff_arrival_at_basement(x):
  x = str(x).replace("\r\n", "")
  x = x.replace(" ", "")
  if len(x) == 4: x = x + "/" + x
  if len(x) == 5: x = x + "/" + x[:4]
  if len(x) == 10: x = x[1:]
  return pd.Series([x])

df[['takeoff_arrival_at_basement']] = df.takeoff_arrival_at_basement.apply(lambda x:fit_takeoff_arrival_at_basement(x))
df = df[df.takeoff_arrival_at_basement.notnull()].query("takeoff_arrival_at_basement != '×'")
df = df.query("takeoff_arrival_at_basement != ''").query("takeoff_arrival_at_basement != '2048/221'").query("takeoff_arrival_at_basement != '2037/9일2033'")
df['takeoff_arrival_at_basement'].unique()

df["transfer_delay"] = df.registration_takeoff.apply(lambda x:x.split("/")[1]) + '/' + df.lz_arrival
df["transfer_delay"].unique()
df[['transfer_delay']] = df.transfer_delay.apply(lambda x:diff_delay(x))

def fit_hoist(x):
  if x in ['호이스트', 'UH-60', '추가임무\r\n연속실시']: return pd.Series([1])
  return pd.Series([0])

df[['hoist']] = df.hoist.apply(lambda x:fit_hoist(x))

def fit_weather_deterioration(x):
  x = str(x)
  if "계기" in x: x = "계기"
  if "악화" in x: x = "악화"
  else: x = "정상"
  return pd.Series([x])

df[['weather_deterioration']] = df.weather_deterioration.apply(lambda x:fit_weather_deterioration(x))

def fit_nfl(x):
  x = str(x)
  if "긴급" in x: x = "긴급"
  if "NFL" in x: x = "NFL"
  else: x = "정상"
  return pd.Series([x])

df[['NFL']] = df.NFL.apply(lambda x:fit_nfl(x))
```

```python
from statistics import mean
from sklearn.model_selection import cross_validate, RepeatedStratifiedKFold
from imblearn.ensemble import BalancedRandomForestClassifier

model_rf = RandomForestClassifier(random_state=1, n_jobs=-1)
model_brf = BalancedRandomForestClassifier(n_estimators=150, random_state=2, n_jobs=-1)
model_logr = LogisticRegression(random_state=1, n_jobs=-1, multi_class='multinomial')
model_xgb = XGBClassifier(random_state=1, n_jobs=-1)
model_gbr = GradientBoostingClassifier(random_state=1)
model_cat = CatBoostClassifier(random_state=1, verbose=0)

models = []
models.append(('LR',model_logr))
models.append(('RF',model_rf))
models.append(('BRF', model_brf))
models.append(('GBR',model_gbr))
models.append(('XGB',model_xgb))
models.append(('CAT',model_cat))
```

```python
scaler = StandardScaler()
onehot = OneHotEncoder(handle_unknown='ignore', sparse=False)
cv = StratifiedKFold(n_splits=5, random_state=1, shuffle=True)
feature = SelectFromModel(model_rf, threshold=0.001)
ct = ColumnTransformer([('onehot', onehot, cat_var), ('scaler', scaler, con_var)], remainder='passthrough', n_jobs=-1)
results = []
names = []

from mlxtend.plotting import plot_confusion_matrix
for name, model in models:
    model.fit(X_train, y_train)
    scores = cross_val_score(model, X_train, y_train, scoring='accuracy', cv=cv, n_jobs=-1, verbose=0)
    names.append(name)
    results.append(scores)
    print("model %s accuracy: %.4f variance: %.4f"%(name, np.mean(scores), np.std(scores)))
    y_pred = model.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    print(classification_report(y_test, y_pred))
    cm = confusion_matrix(y_test, y_pred)
    fig, ax = plot_confusion_matrix(conf_mat=cm)
    plt.xlabel('Predictions', fontsize=18)
    plt.ylabel('Actuals', fontsize=18)
    plt.title('Confusion Matrix', fontsize=18)
    plt.show()
```

```python
from imblearn.over_sampling import SMOTE

# Use SMOTE to oversample the minority clss
# SMOTE takes in consideration k_neighbors=5 by default. If the datast_sample requires it to, the model has to take fewer k_neighbors, just like we did here
# oversample = SMOTE(k_neighbors=3)
oversample = SMOTE(k_neighbors=6)
over_X, over_y = oversample.fit_resample(X, y)

over_X_train, over_X_test, over_y_train, over_y_test = train_test_split(over_X, over_y, test_size=0.4, stratify=over_y)

# Build SMOTE SRF model
SMOTE_SRF = RandomForestClassifier(n_estimators=150, random_state=0)

cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
scoring = ('accuracy', 'precision', 'recall')

# Evaluate SMOTE SRF model
scores = cross_validate(SMOTE_SRF, over_X, over_y, scoring=scoring, cv=cv)

#Train SMOTE SRF
SMOTE_SRF.fit(over_X_train, over_y_train)
#SMOTE SRF prediction result
smote_y_pred = SMOTE_SRF.predict(over_X_test)

print("Accuracy: ", accuracy_score(over_y_test, smote_y_pred))
print("Precision: ", precision_score(over_y_test, smote_y_pred))
print("Recall:", recall_score(over_y_test, smote_y_pred))
saveTrees(SMOTE_SRF, "img/SMOTE/")


# Create the confusion matrix
smote_cm = confusion_matrix(over_y_test, smote_y_pred)
ConfusionMatrixDisplay(confusion_matrix=smote_cm).plot()
#saveConfusionMatrix(smote_cm, "img/SMOTE/")

# View the classification report for test data and predictions
print(classification_report(over_y_test, smote_y_pred))
```
